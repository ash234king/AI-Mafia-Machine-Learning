```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dfx=pd.read_csv(r"C:\Users\Hosiyar Singh\Documents\python machine learning\linearX.csv")
dfy=pd.read_csv(r"C:\Users\Hosiyar Singh\Documents\python machine learning\linearY.csv")
x=dfx.values
y=dfy.values

x=x.reshape((-1,1))
print(x)
y=y.reshape((-1,1))
print(x.shape)
print(y.shape)
```

```python
print(x,type(x))
```

```python
# plotting 
plt.scatter(x,y,color='red')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(x,y) #it accepts a 2D list so we make x and y again a 2D list
```

```python
output=model.predict(x)
```

```python
bias=model.intercept_
coeff=model.coef_
print(bias)
print(coeff)
```

```python
model.score(x,y)
```

```python
plt.scatter(x,y,label='data')
plt.plot(x,output,color='black', label='prediction')
plt.legend()
plt.show()
```
