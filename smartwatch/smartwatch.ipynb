```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=pd.read_csv(r"C:\Users\Hosiyar Singh\Documents\python machine learning\Linear_X_Train.csv")
y=pd.read_csv(r"C:\Users\Hosiyar Singh\Documents\python machine learning\Linear_Y_Train.csv")
x=x.values
y=y.values
x=x.reshape((-1,))
y=y.reshape((-1,))
print(x.shape,y.shape)
```

```python
plt.scatter(x,y)
plt.show()
```

```python
x=(x-x.mean())/x.std()
plt.scatter(x,y)
plt.show()
```

```python
def hypothesis(x,theta):
    return theta[0]+theta[1]*x
```

```python
def error(x,y,theta):
    error=0.0
    for i in range(x.shape[0]):
        hx=hypothesis(x[i],theta)
        error+=(hx-y[i])**2
    return error
```

```python
def gradient(x,y,theta):
    grad=np.zeros((2,))
    for i in range(x.shape[0]):
        hx=hypothesis(x[i],theta)
        grad[0]+=(hx-y[i])
        grad[1]+=(hx-y[i])*x[i]
    return grad/x.shape[0]
```

```python
def gradientdescent(x,y,learning_rate=0.01):
    theta=np.zeros((2,))
    error_list=[]
    grad_list=[]
    for i in range(1000):
        grad=gradient(x,y,theta)
        err=error(x,y,theta)
        error_list.append(err)
        grad_list.append(grad)
        theta[0]-=learning_rate*grad[0]
        theta[1]-=learning_rate*grad[1]
    return theta,error_list,grad_list
```

```python
final_theta,error_list,grad_list=gradientdescent(x,y)
```

```python
plt.plot(error_list)
plt.show()
```
